```
# Signal Generation
# matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
mean = 0
std = 1 
num_samples = 1000
t = np.random.normal(mean, std, size=num_samples)


x_db = 10/ * np.log10(t)
plt.subplot(3,1,3)
plt.plot(t)
plt.title('random noise')
plt.ylabel('Amplitude')
plt.xlabel('Time (s)')
plt.show()
```
